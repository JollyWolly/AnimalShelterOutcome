```python
import pandas as pd
import numpy as np

import datetime
from datetime import datetime

import sklearn

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

import warnings
warnings.filterwarnings('ignore')
```

```python
pet_df = pd.read_csv("Animal_Shelter_Intake_and_Outcome.csv")
pet_df.shape
```

```python
pet_df.head()
```

```python
pet_df = pet_df.drop(['Name','Impound Number','Kennel Number','Animal ID',
                      'Outcome Date','Intake Subtype','Outcome Subtype',
                      'Outcome Condition','Intake Jurisdiction','Outcome Jurisdiction',
                      'Outcome Zip Code','Location','Count'
                     ], axis=1).dropna()
pet_df.head()
```

```python
print("---INPUT CATEGORICAL---")

print("TYPES : ")
print(pet_df['Type'].unique().tolist())

print("\nSEXES : ")
print(pet_df['Sex'].unique().tolist())

print("\nSIZES : ")
print(pet_df['Size'].unique().tolist())

print("\nINTAKE TYPES : ")
print(pet_df['Intake Type'].unique().tolist())

print("\nINTAKE CONDITIONS : ")
print(pet_df['Intake Condition'].unique().tolist())

print("\nBREEDS : ")
print(pet_df['Breed'].unique().tolist())

print("\nCOLORS : ")
print(pet_df['Color'].unique().tolist())
```

```python
print("---OUTPUT CATEGORICAL---")

print("OUTCOME TYPES : ")
print(pet_df['Outcome Type'].unique().tolist())
```

```python
# Remove other from types
pet_df = pet_df.loc[pet_df['Type'] != 'OTHER']
```

```python
# Change Sex
pet_df['Sex'].value_counts()
```

```python
# Unknown Sex is a small sample here, so I'm going to remove it
pet_df = pet_df.loc[pet_df['Sex'] != 'Unknown'].reset_index(drop=True)

pet_df['Fixed'] = pet_df['Sex'].map({'Neutered' : True, 'Spayed' : True,
                                    'Male' : False, 'Female' : False})
pet_df['Sex'] = pet_df['Sex'].map({'Neutered' : 'Male', 'Male' : 'Male',
                                   'Spayed' : 'Female', 'Female' : 'Female'})
pet_df.head()
```

```python
pet_df['Size'] = pet_df['Size'].map({'KITTN' : 0, 'PUPPY' : 0, 
                                     'TOY' : 1, 'SMALL' : 1,
                                     'MED' : 2, 'LARGE' : 3,
                                     'X-LRG' : 4})
pet_df['Size'].value_counts()
```

```python
f = lambda x: datetime.strptime(x, '%m/%d/%Y')
bDt = pet_df["Date Of Birth"].apply(f)
iDt = pet_df["Intake Date"].apply(f)

pet_df["Age"] = iDt-bDt
pet_df["Age"] = pet_df["Age"].dt.days + pet_df["Days in Shelter"]

pet_df.head()
```

```python
len(pet_df['Breed'].unique())
```

```python
def splitCol(x):
    if '/' in x:
        return x.split('/',1)
    return [x, 'None']
```

```python
brd_ls = pet_df.apply(lambda b : splitCol(b['Breed']), axis=1)

pet_df['Breed Primary'] = [e[0] for e in brd_ls]
pet_df['Breed Secondary'] = [e[1] for e in brd_ls]
```

```python
clr_ls = pet_df.apply(lambda c : splitCol(c['Color']), axis=1)

pet_df['Color Primary'] = [e[0] for e in clr_ls]
pet_df['Color Secondary'] = [e[1] for e in clr_ls]
```

```python
pet_df.head()
```

```python
pet_df['Outcome Type'].unique()
```

```python
pet_df['Outcome Type'] = pet_df['Outcome Type'].map({'ADOPTION':'ADOPT', 'RETURN TO OWNER':'RTO',
                                                    'TRANSFER':'TRANS', 'DISPOSAL':'EUTH', 'EUTHANIZE':'EUTH',
                                                    'DIED':'DIED', 'RTOS':'RTO', 'ESCAPED/STOLEN':'ESCP'})
```

```python
pet_df = pet_df.drop(['Breed','Color','Date Of Birth','Intake Date'], axis=1)
pet_df.head()
```

```python
in_df = pet_df.drop(['Outcome Type'], axis=1)
out_df = pet_df['Outcome Type']
in_df.head()
```

```python
pt_inT = pd.get_dummies(in_df['Intake Type'])
pt_inC = pd.get_dummies(in_df['Intake Condition'])
```

```python
pt_clr = pd.get_dummies(in_df['Color Primary'])
pt_brd = pd.get_dummies(in_df['Breed Primary'])

for i in range(0, len(in_df['Type'])):
    c = in_df['Color Secondary'][i]
    b = in_df['Breed Secondary'][i]
    
    if c != 'None':
        if c not in pt_clr:
            pt_clr[c] = 0
        pt_clr[c][i] = 2
        
    if b != 'None':
        if b not in pt_brd:
            pt_brd[b] = 0
        pt_brd[b][i] = 2
```

```python
in_bin = pd.DataFrame()
in_bin['Type(dog=1)'] = in_df.apply(lambda a : int(a['Type'] == 'DOG'), axis=1)
in_bin['Sex(male=1)'] = in_df.apply(lambda b : int(b['Sex'] == 'Male'), axis=1)
in_bin['Fixed']       = in_df.apply(lambda c : int(c['Fixed']), axis=1)
in_bin['Age(days)']   = in_df['Age']
in_bin['Days Held'] = in_df['Days in Shelter']

in_bin = pd.concat([in_bin, pt_inT, pt_inC, pt_clr, pt_brd], axis=1)
in_bin.head()
```

```python
out_bin = pd.get_dummies(pet_df['Outcome Type'])
out_bin.head()
```

```python
a_train, a_test, b_train, b_test = train_test_split(in_bin, out_bin, test_size = 0.33, random_state = 42)
```

```python
rfc = RandomForestClassifier(n_estimators=100, random_state=0)

rfc.fit(a_train, b_train)

b_pred = rfc.predict(a_test)


print('Model accuracy score with 100 decision-trees : {0:0.4f}'. format(accuracy_score(b_test, b_pred)))
```

```python
feature_scores = pd.Series(rfc.feature_importances_, index=a_train.columns).sort_values(ascending=False)

feature_scores
```

```python
from sklearn.metrics import classification_report

print(classification_report(b_test, b_pred))
```
